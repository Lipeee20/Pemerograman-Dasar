```python
# Tampilan menu sekali di awal
print("=" * 45)
print("        SELAMAT DATANG DI SIPIT CAFE")
print("=" * 45)
print("Menu Makanan & Minuman:")
print("1. Milo Giant            - Rp 18.000")
print("2. Air Putih             - Rp 5.000")
print("3. Aneka Jus Buah        - Rp 12.000")
print("4. Roti Gembul           - Rp 25.000")
print("5. Mie Goreng Seafood    - Rp 20.000")
print("6. Ayam Bakar            - Rp 18.000")
print("7. Nasi Goreng           - Rp 15.000")
print("=" * 45)

# Dictionary menu
menu = {
    1: ("Milo Giant", 18000),
    2: ("Air Putih", 5000),
    3: ("Aneka Jus Buah", 12000),
    4: ("Roti Gembul", 25000),
    5: ("Mie Goreng Seafood", 20000),
    6: ("Ayam Bakar", 18000),
    7: ("Nasi Goreng", 15000)
}

# Input jumlah item yang ingin dipesan
pesanan_list = []
total_harga = 0
jumlah_item = int(input("Berapa jenis menu yang ingin Anda pesan? "))

for i in range(jumlah_item):
    pilihan = int(input(f"Pilih menu ke-{i+1} (1-7): "))

    if pilihan in menu:
        nama, harga = menu[pilihan]
        jumlah = int(input(f"Masukkan jumlah untuk {nama}: "))
        subtotal = harga * jumlah
        pesanan_list.append((nama, jumlah, harga, subtotal))
        total_harga += subtotal
        print(f"‚úÖ {jumlah}x {nama} ditambahkan (Subtotal: Rp {subtotal:,})")
    else:
        print("‚ùå Menu tidak valid, dilewati.")

# Hitung diskon
if total_harga >= 100000:
    diskon = 0.70
elif total_harga >= 50000:
    diskon = 0.50
elif total_harga >= 20000:
    diskon = 0.30
else:
    diskon = 0

potongan = total_harga * diskon
total_bayar = total_harga - potongan

# Cetak struk
print("*" * 50)
print("                 STRUK PEMBAYARAN")
print("*" * 50)

for nama, jumlah, harga, subtotal in pesanan_list:
    print(f"{jumlah}x {nama:<20} - Rp {subtotal:,}")

print("-" * 50)
print(f"Total Sebelum Diskon : Rp {total_harga:,}")

if diskon > 0:
    print(f"Diskon               : {int(diskon * 100)}%")
    print(f"Potongan             : Rp {potongan:,.0f}")
else:
    print("Diskon               : Tidak ada")

print(f"Total Bayar          : Rp {total_bayar:,.0f}")
print("-" * 50)

uang_dibayar = int(input("Cash                 : Rp "))
if uang_dibayar >= total_bayar:
    kembalian = uang_dibayar - total_bayar
    print(f"Kembalian            : Rp {kembalian:,.0f}")
else:
    kekurangan = total_bayar - uang_dibayar
    print(f"‚ùå Uang kurang Rp {kekurangan:,.0f}. Harap bayar sesuai total.")

print("*" * 50)
print("       Terima kasih, selamat menikmati üòä")
print("*" * 50)
```
