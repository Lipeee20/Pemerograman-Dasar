```python
all_df["order_month_year"] = all_df["order_date"].dt.strftime("%Y-%m")
all_df.head()
```

```python
monthly_sales_revenue_df = all_df.groupby("order_month_year").agg({
    "quantity_x": "sum",
    "total_price": "sum"
}).reset_index()
monthly_sales_revenue_df.rename(columns={'quantity_x': 'monthly_sales_quantity', 'total_price': 'monthly_revenue'}, inplace=True)
monthly_sales_revenue_df.head()
```

```python
plt.figure(figsize=(15, 6))

# Plot for Monthly Sales Quantity
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.plot(monthly_sales_revenue_df["order_month_year"], monthly_sales_revenue_df["monthly_sales_quantity"], marker='o', linestyle='-', color='skyblue')
plt.title('Monthly Sales Quantity Trend')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales Quantity')
plt.xticks(rotation=45)
plt.grid(True)

# Plot for Monthly Revenue
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.plot(monthly_sales_revenue_df["order_month_year"], monthly_sales_revenue_df["monthly_revenue"], marker='o', linestyle='-', color='lightcoral')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month-Year')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
product_performance_df = sales_products_df.groupby("product_name").agg({
    "quantity_x": "sum",
    "total_price": "sum"
}).reset_index()
product_performance_df.head()
```
