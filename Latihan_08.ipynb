```python
import pandas as pd
from dateutil import parser

# 1) Ubah semua jadi string
LIPEEE_df['tanggal_daftar'] = LIPEEE_df['tanggal_daftar'].astype(str)

# 2) Hapus karakter aneh atau whitespace tersembunyi
LIPEEE_df['tanggal_daftar'] = LIPEEE_df['tanggal_daftar'].str.strip()

# 3) Parsing aman
def smart_parse(x):
    try:
        return parser.parse(x)
    except:
        return pd.NaT

LIPEEE_df['tanggal_daftar'] = LIPEEE_df['tanggal_daftar'].apply(smart_parse)

# 4) Format akhir
LIPEEE_df['tanggal_daftar'] = LIPEEE_df['tanggal_daftar'].dt.strftime('%Y-%m-%d')
```

Penjelasan Langkah-langkahnya

- Mengubah kolom jadi string

  Kita ubah dulu semua isi tanggal_daftar menjadi tipe string. Tujuannya supaya format data apa pun (angka, teks, campuran) aman untuk diproses tanpa error.

- Membersihkan whitespace
  
  Setelah itu kita hapus spasi berlebih atau karakter tersembunyi dengan .str.strip(). Kadang tanggal gagal diproses hanya karena ada spasi di depan/belakang.

- Parsing memakai smart_parse
  
  Fungsi smart_parse mencoba membaca tanggal apa pun, misalnya 24 Nov 2024, 2024/03/10, 09-17-2024, dll. Kalau formatnya benar â†’ dikonversi menjadi datetime. Kalau formatnya kacau â†’ otomatis jadi NaT (missing). Jadi parsing aman tanpa bikin error.

- Menyeragamkan format ke YYYY-MM-DD
  
  Setelah semua tanggal berhasil dibaca, kita ubah tampilannya menjadi satu format standar YYYY-MM-DD. Ini membuat kolom tanggal jadi rapi dan mudah dipakai untuk sorting, filtering, atau analisis.
